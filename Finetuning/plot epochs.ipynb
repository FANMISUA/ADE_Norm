```python
import matplotlib.pyplot as plt

def read_losses_from_file(filename):
    """Read training losses from a text file and return epoch and loss lists."""
    epochs = []
    losses = []
    with open(filename, 'r') as file:
        for line in file:
            # Split line to get epoch number and loss value
            parts = line.strip().split(':')
            epoch = int(parts[0].split()[1])
            loss = float(parts[1].strip())
            epochs.append(epoch)
            losses.append(loss)
    return epochs, losses

# Read data from files
epochs1, losses1 = read_losses_from_file('./smm4h_top3_training_losses.txt')
epochs2, losses2 = read_losses_from_file('./cadec_top3_training_losses.txt')
epochs3, losses3 = read_losses_from_file('./combine_top3_training_losses.txt')

# Ensure epochs are the same across files# Assuming they are aligned; if not, you might need to handle different lengths or align epochs# Plotting training losses
plt.figure(figsize=(12, 8))

# Plot each file's data with a different color
plt.plot(epochs1, losses1, '-o', color='b', label='smm4h top3')
plt.plot(epochs2, losses2, '-o', color='g', label='cadec top3')
plt.plot(epochs3, losses3, '-o', color='r', label='combine top3')

# Customize font sizes
plt.xlabel('Epoch', fontsize=14)
plt.ylabel('Training Loss', fontsize=14)
plt.title('Training Loss vs. Epochs', fontsize=16)
plt.legend(fontsize=12)
plt.grid(True)

# Save and show the plot
plt.savefig('top6_training_loss_comparison.png')
plt.show()


# Read data from files
epochs4, losses4 = read_losses_from_file('./smm4h_top6_training_losses.txt')
epochs5, losses5 = read_losses_from_file('./cadec_top6_training_losses.txt')
epochs6, losses6 = read_losses_from_file('./combine_top6_training_losses.txt')

# Ensure epochs are the same across files# Assuming they are aligned; if not, you might need to handle different lengths or align epochs# Plotting training losses
plt.figure(figsize=(12, 8))

# Plot each file's data with a different color
plt.plot(epochs4, losses4, '-o', color='b', label='smm4h top6')
plt.plot(epochs5, losses5, '-o', color='g', label='cadec top6')
plt.plot(epochs6, losses6, '-o', color='r', label='combine top6')

# Customize font sizes
plt.xlabel('Epoch', fontsize=14)
plt.ylabel('Training Loss', fontsize=14)
plt.title('Training Loss vs. Epochs', fontsize=16)
plt.legend(fontsize=12)
plt.grid(True)

# Save and show the plot
plt.savefig('top6_training_loss_comparison.png')
plt.show()


# Read data from files
epochs7, losses7 = read_losses_from_file('./smm4h_all_training_losses.txt')
epochs8, losses8 = read_losses_from_file('./cadec_all_training_losses.txt')
epochs9, losses9 = read_losses_from_file('./combine_all_training_losses.txt')

# Ensure epochs are the same across files# Assuming they are aligned; if not, you might need to handle different lengths or align epochs# Plotting training losses
plt.figure(figsize=(12, 8))

# Plot each file's data with a different color
plt.plot(epochs7, losses7, '-o', color='b', label='smm4h all')
plt.plot(epochs8, losses8, '-o', color='g', label='cadec all')
plt.plot(epochs9, losses9, '-o', color='r', label='combine all')

# Customize font sizes
plt.xlabel('Epoch', fontsize=14)
plt.ylabel('Training Loss', fontsize=14)
plt.title('Training Loss vs. Epochs', fontsize=16)
plt.legend(fontsize=12)
plt.grid(True)

# Save and show the plot
plt.savefig('all_training_loss_comparison.png')
plt.show()
```
